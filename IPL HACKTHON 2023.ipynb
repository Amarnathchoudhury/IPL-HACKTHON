# 10. To plot the most consistent batsmen among top 10 run getters, what plotting technique we use?

```python
top_10_runs = data.groupby('batsman')['batsman_runs'].sum().nlargest(10).index
top_10_data = data[data['batsman'].isin(top_10_runs)]


plt.figure(figsize=(12, 6))
sns.lineplot(data=top_10_data, x='batsman', y='batsman_runs', estimator='mean', marker='o')
plt.title('Consistency of Top 10 Run-Getters')
plt.xlabel('Batsman')
plt.ylabel('Average Runs per Match')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# 11. In 2019 which batsman scored highest number of runs by hitting 6's and 4's?

```python
data_2019 = merged_data[merged_data['season'] == 2019]
runs_by_type = data_2019.groupby(['batsman', 'batsman_runs'])['batsman_runs'].count().unstack(fill_value=0)
runs_by_type['total_runs'] = runs_by_type[4] * 4 + runs_by_type[6] * 6
highest_runs_batsman = runs_by_type['total_runs'].idxmax()
```

# 13.  What is the strike rate of Kohli in 2016
